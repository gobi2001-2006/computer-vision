```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("pic.jpeg", cv2.IMREAD_GRAYSCALE)

sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

prewitt_x = cv2.filter2D(image, cv2.CV_64F, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(image, cv2.CV_64F, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
prewitt_combined = cv2.magnitude(prewitt_x, prewitt_y)

roberts_x = cv2.filter2D(image, cv2.CV_64F, np.array([[1, 0], [0, -1]]))
roberts_y = cv2.filter2D(image, cv2.CV_64F, np.array([[0, 1], [-1, 0]]))
roberts_combined = cv2.magnitude(roberts_x, roberts_y)


canny = cv2.Canny(image, 50, 150)

# Plot results
titles = ['Original Image', 'Sobel', 'Prewitt', 'Roberts', 'Canny']
images = [image, sobel_combined, prewitt_combined, roberts_combined,  canny]

plt.figure(figsize=(12, 6))
for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread("pic.jpeg", cv2.IMREAD_GRAYSCALE)

# Apply Gaussian Blur to smooth the image
blurred = cv2.GaussianBlur(image, (3, 3), 0)

# Apply Laplacian operator
log = cv2.Laplacian(blurred, cv2.CV_64F)

# Convert the result to uint8 type


# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title("Original Image")
plt.subplot(1,2,2), plt.imshow(log, cmap='gray'), plt.title("Laplacian of Gaussian (LoG)")
plt.show()
```

```python
# Apply Gaussian Blur with two different sigma values
gaussian1 = cv2.GaussianBlur(image, (5, 5), 1)  # Small blur
gaussian2 = cv2.GaussianBlur(image, (5, 5), 2)  # Larger blur

# Compute the Difference of Gaussian (DoG)
dog = gaussian1 - gaussian2

# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title("Original Image")
plt.subplot(1,2,2), plt.imshow(dog, cmap='gray'), plt.title("Difference of Gaussian (DoG)")
plt.show()
```
